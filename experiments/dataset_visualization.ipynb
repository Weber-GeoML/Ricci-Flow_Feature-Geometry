# Synthetic Datasets

```python
X_circles, y_circles = DatasetFactory.make_circles(noise=0.04)

X_4circles, y_4circles = DatasetFactory.make_4circles(n=10000)

X_cylinder, y_cylinder = DatasetFactory.make_cylinders(n=10000)

X_torus, y_torus = DatasetFactory.make_tori(10000, center2=(2,0,0))

# Create the figure
fig = plt.figure(figsize=(24, 6))
#plt.suptitle("Synthetic Datasets")
gs = gridspec.GridSpec(1, 5, figure=fig, wspace=0.4)

# First plot (2D)
ax1 = fig.add_subplot(gs[0])
ax1.scatter(x=X_circles[:, 0], y=X_circles[:, 1], c=y_circles)
ax1.set_title("Syn-I", fontsize=25)
ax1.set_aspect('equal')  # ensure equal scaling

# Second plot (2D)
ax2 = fig.add_subplot(gs[1])
ax2.scatter(x=X_4circles[:, 0], y=X_4circles[:, 1], c=y_4circles)
ax2.set_title("Syn-II", fontsize=25)
ax2.set_aspect('equal')  # ensure equal scaling

# Third plot (3D)
ax3 = fig.add_subplot(gs[2], projection='3d')
ax3.scatter(X_cylinder[:, 0], X_cylinder[:, 1], X_cylinder[:, 2], c=y_cylinder)
ax3.set_title("Syn-III", fontsize=25)
ax3.set_box_aspect([1, 1, 1])  # equal aspect ratio

# Foruth plot (3D)
ax4 = fig.add_subplot(gs[3], projection='3d')
ax4.scatter(X_torus[:, 0], X_torus[:, 1], X_torus[:, 2], c=y_torus)
ax4.set_title("Syn-IV", fontsize=25)
ax4.set_box_aspect([1, 1, 1])


ax1.set_xticks([])
ax1.set_yticks([])

ax2.set_xticks([])
ax2.set_yticks([])

ax3.set_xticklabels([])
ax3.set_yticklabels([])
ax3.set_zticklabels([])

ax4.set_xticklabels([])
ax4.set_yticklabels([])
ax4.set_zticklabels([])

plt.savefig("../plots/synthetic_datasets.png", bbox_inches='tight', dpi=300)
plt.show()
```

# Real-World Datasets

```python
X_MNIST_69, _, _, _ = DatasetFactory.load_MNIST((6,9), device='cpu')
X_MNIST_17, _, _, _ = DatasetFactory.load_MNIST((1,7), device='cpu')
X_fMNIST_apperal, _, _, _ = DatasetFactory.load_fMNIST((3,6), device='cpu')
X_fMNIST_sneakers, _, _, _ = DatasetFactory.load_fMNIST((5,7), device='cpu')
X_CIFAR, _, _, _ = DatasetFactory.load_CIFAR(classes=(0,1), device='cpu')

plt.figure(figsize=(12, 6))
for i in range(10):
    plt.subplot(5, 10, i + 1)
    plt.imshow(X_MNIST_69[i].squeeze(), cmap="gray")
    plt.axis("off")
for i in range(10):
    plt.subplot(5, 10, 10+i + 1)
    plt.imshow(X_MNIST_17[i].squeeze(), cmap="gray")
    plt.axis("off")
for i in range(10):
    plt.subplot(5, 10, 20+i + 1)
    plt.imshow(X_fMNIST_apperal[i].squeeze(), cmap="gray")
    plt.axis("off")
for i in range(10):
    plt.subplot(5, 10, 30+i + 1)
    plt.imshow(X_fMNIST_sneakers[i].squeeze(), cmap="gray")
    plt.axis("off")
for i in range(10):
    plt.subplot(5, 10, 40+i + 1)
    plt.imshow(X_CIFAR[i])
    plt.axis("off")
plt.tight_layout()
plt.savefig('../plots/real_world_datasets.png', bbox_inches="tight", dpi=300)
```
